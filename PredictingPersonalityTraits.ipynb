# Extraction of videos and images

The following 3 cells are used to extract the videos from the zipped archives that came in the dataset and then frames from the videos and store them seperate folders. One folder per video is used.

- The first cell is for training files.

- The second cell for validation files.

- The third cell is for test files.


```python
'''
Extract all the 6 training zipped files into a folder and then run this script
'''
import cv2
import numpy as np
import os
import zipfile

for i in range(1,76):
    ## Runnin a loop throught all the zipped training file to extract all video and then extract 100 frames from each.
    if i<10:
        zipfilename = 'training80_0'+str(i)+'.zip'
    else:
        zipfilename = 'training80_'+str(i)+'.zip'
    ## Accessing the zipfile i
    archive = zipfile.ZipFile('data/'+zipfilename, 'r')
    zipfilename = zipfilename.split('.zip')[0]

    ##Extracting all videos in it and saving it all to the new folder with same name as zipped one
    archive.extractall('unzippedData/'+zipfilename)
    
    ## Running a loop over all the videos in the zipped file and extracting 100 frames from each
    for file_name in archive.namelist():
        cap = cv2.VideoCapture('unzippedData/'+zipfilename+'/'+file_name)

        file_name=(file_name.split('.mp4'))[0]
        ## Creating folder to save all the 100 frames from the video
        try:
            if not os.path.exists('ImageData/trainingData/'+file_name):
                os.makedirs('ImageData/trainingData/'+file_name)
        except OSError:
            print ('Error: Creating directory of data')

        ## Setting the frame limit to 100
        #cap.set(cv2.CAP_PROP_FRAME_COUNT, 101)
        length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        print( length )
        #length=101
        count=0
        framecount = 1
        ## Running a loop to each frame and saving it in the created folder
        while(cap.isOpened()):
            if length==count:
                break
            ret, frame = cap.read()
            if frame is None:
                continue
            
            if count%4 == 0 and framecount<101:
                framecount = framecount+1
                #print('count', count,'framecount', framecount)
                ## Resizing it to 256*256 to save the disk space and fit into the model
                frame = cv2.resize(frame,(256, 256), interpolation = cv2.INTER_CUBIC)
                # Saves image of the current frame in jpg file
                name = 'ImageData/trainingData/'+str(file_name)+'/frame' + str(count) + '.jpg'
                cv2.imwrite(name, frame)
        
            count+=1
            
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        ## Print the file which is done
        print(zipfilename, ':', file_name)
                
```

# Storing image datasets into TFRecords

Due to large size of the datasets and to enable ease of handling data one TFRecord each is used to store the training, validation and test dataset. Each image is stored as a byte feature along with the corresponding ground truth values as another byte feature in an Example(tf.example message). These examples are serially written into a TFRecord.

The next three cells are used to do the same for training, validation and test.

```python
from   random import shuffle
import glob
import pandas as pd
import tensorflow as tf
import sys
import numpy as np
import PIL.Image
import cv2
import os

def load_image(addr):
    # read an image and resize to (224, 224)
    # cv2 load images as BGR, convert it to RGB
    img = np.array(PIL.Image.open(addr).resize((224,224), PIL.Image.ANTIALIAS))
    img = img.astype(np.uint8)
    return img

def _float_feature(value):
    return tf.train.Feature(float_list=tf.train.FloatList(value=[value]))

def _bytes_feature(value):
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))


df = pd.read_csv('./data/training_gt.csv')
NUM_VID = len(df)
addrs = []
labels = []

#duplicates the label values for number of frames in the video and obtains the path of each image(frame of video)
for i in range(NUM_VID):
    filelist=glob.glob('ImageData/trainingData/'+(df['VideoName'].iloc[i]).split('.mp4')[0]+'/*.jpg')
    addrs+=filelist
    listofImages = os.listdir('ImageData/trainingData/'+(df['VideoName'].iloc[i]).split('.mp4')[0]) # dir is your directory path
    number_files = len(listofImages)
    print (number_files) 
    labels+=[np.array(df.drop(['VideoName'], 1, inplace=False).iloc[i]).astype(np.float32)]*number_files

c = list(zip(addrs, labels))
shuffle(c)
train_addrs, train_labels = zip(*c)
train_filename = 'train_full.tfrecords'  # address to save the TFRecords file
# open the TFRecords file
writer = tf.io.TFRecordWriter(train_filename)

for i in range(len(train_addrs)):
    # print how many images are saved every 1000 images
    if not i % 1000:
        print ('Train data: {}/{}'.format(i, len(train_addrs)))
        sys.stdout.flush()
    # Load the image
    img = load_image(train_addrs[i])
    # print img[0,0:5]
    # break
    # print img.shape
    label = train_labels[i]
    # print label.shape
    # print label
    # Create a feature
    feature = {'train/label': _bytes_feature(tf.compat.as_bytes(label.tostring())),
               'train/image': _bytes_feature(tf.compat.as_bytes(img.tostring()))}
    # Create an example protocol buffer
    example = tf.train.Example(features=tf.train.Features(feature=feature))
   
    # Serialize to string and write on the file
    writer.write(example.SerializeToString())


writer.close()
sys.stdout.flush()

print (len(train_addrs)), "training images saved.. "
```

# DAN - Descriptor Aggregator Network

The following cell defines a class which corresponds to the neural network architecture used for facial features part of the model. 

The DAN is a modified CNN where the last fully connected layers are replaced by two branches, one containing a max pool layer and another an average pool layer each followed by L2-normalization. The output of the L2-normalization on both branches is concatenated and that is fed as input to the final sigmoid regression layer which predicts the personality trait values for the Facial Features modality.

```python
########################################################################################
#                                                                                      #
# VGG16 implementation in TensorFlow                                                   #
# Details:                                                                             #
# http://www.cs.toronto.edu/~frossard/post/vgg16/                                      #
#                                                                                      #
# Model from https://gist.github.com/ksimonyan/211839e770f7b538e2d8#file-readme-md     #
# Weights from Caffe converted using https://github.com/ethereon/caffe-tensorflow      #
########################################################################################

import tensorflow as tf
import numpy as np
import warnings
from scipy.io import loadmat
import pickle


warnings.filterwarnings("ignore")




class DAN:
    def __init__(self, imgs, REG_PENALTY=0, preprocess=None):
        self.imgs = imgs
        if preprocess=='vggface':
            self.mean = [129.1862793, 104.76238251, 93.59396362]
        else:
            self.mean = [123.68, 116.779, 103.939]
            
        #calling convolutional layers
        self.convlayers()
        #calling the DAN modification part of CNN
        self.dan_part()
        self.output = tf.nn.sigmoid(self.reg_head, name="output")
        self.cost_reg = REG_PENALTY*tf.reduce_mean(tf.square(self.parameters[-2]))/2
        
    #defining convolutional layers
    def convlayers(self):
        self.parameters = []

        # zero-mean input
        with tf.name_scope('preprocess') as scope:
            mean = tf.constant(self.mean, dtype=tf.float32, shape=[1, 1, 1, 3], name='img_mean')
            images = self.imgs-mean
            print("preprocess")

        # conv1_1
        with tf.name_scope('conv1_1') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 3, 64], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(images, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[64], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv1_1 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv1_1")

        # conv1_2
        with tf.name_scope('conv1_2') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 64, 64], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(self.conv1_1, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[64], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv1_2 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv1_2")

        # pool1
        self.pool1 = tf.nn.max_pool(self.conv1_2,
                               ksize=[1, 2, 2, 1],
                               strides=[1, 2, 2, 1],
                               padding='SAME',
                               name='pool1')

        # conv2_1
        with tf.name_scope('conv2_1') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 64, 128], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(self.pool1, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[128], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv2_1 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv2_1")

        # conv2_2
        with tf.name_scope('conv2_2') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 128, 128], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(self.conv2_1, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[128], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv2_2 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv2_2")

        # pool2
        self.pool2 = tf.nn.max_pool(self.conv2_2,
                               ksize=[1, 2, 2, 1],
                               strides=[1, 2, 2, 1],
                               padding='SAME',
                               name='pool2')

        # conv3_1
        with tf.name_scope('conv3_1') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 128, 256], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(self.pool2, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[256], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv3_1 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv3_1")

        # conv3_2
        with tf.name_scope('conv3_2') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 256, 256], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(self.conv3_1, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[256], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv3_2 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv3_2")

        # conv3_3
        with tf.name_scope('conv3_3') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 256, 256], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(self.conv3_2, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[256], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv3_3 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv3_3")

        # pool3
        self.pool3 = tf.nn.max_pool(self.conv3_3,
                               ksize=[1, 2, 2, 1],
                               strides=[1, 2, 2, 1],
                               padding='SAME',
                               name='pool3')

        # conv4_1
        with tf.name_scope('conv4_1') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 256, 512], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(self.pool3, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[512], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv4_1 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv4_1")

        # conv4_2
        with tf.name_scope('conv4_2') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 512, 512], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(self.conv4_1, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[512], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv4_2 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv4_2")

        # conv4_3
        with tf.name_scope('conv4_3') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 512, 512], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(self.conv4_2, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[512], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv4_3 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv4_3")

        # pool4
        self.pool4 = tf.nn.max_pool(self.conv4_3,
                               ksize=[1, 2, 2, 1],
                               strides=[1, 2, 2, 1],
                               padding='SAME',
                               name='pool4')

        # conv5_1
        with tf.name_scope('conv5_1') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 512, 512], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(self.pool4, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[512], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv5_1 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv5_1")

        # conv5_2
        with tf.name_scope('conv5_2') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 512, 512], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(self.conv5_1, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[512], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv5_2 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv5_2")

        # conv5_3
        with tf.name_scope('conv5_3') as scope:
            kernel = tf.Variable(tf.truncated_normal([3, 3, 512, 512], dtype=tf.float32,
                                                     stddev=1e-1), name='weights')
            conv = tf.nn.conv2d(self.conv5_2, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.Variable(tf.constant(0.0, shape=[512], dtype=tf.float32),
                                 trainable=True, name='biases')
            out = tf.nn.bias_add(conv, biases)
            self.conv5_3 = tf.nn.relu(out, name=scope)
            self.parameters += [kernel, biases]
            print("inside conv5_3")

        # MaxPool5
        self.maxpool5 = tf.nn.max_pool(self.conv5_3,
                               ksize=[1, 2, 2, 1],
                               strides=[1, 2, 2, 1],
                               padding='SAME',
                               name='maxpool5')

        # AvgPool5
        self.avgpool5 = tf.nn.avg_pool(self.conv5_3,
                               ksize=[1, 2, 2, 1],
                               strides=[1, 2, 2, 1],
                               padding='SAME',
                               name='avgpool5')

    #replacing fully connected layers of tradional CNN with max-pool and avg-pool layers here
    def dan_part(self):
        
        with tf.name_scope('reg_head') as scope:
            shape = int(2*int(np.prod(self.maxpool5.get_shape()[1:])))
            # shape = int(np.prod(self.pool5.get_shape()[1:]))
            fc1w = tf.Variable(tf.truncated_normal([shape, 5],
                                                         dtype=tf.float32,
                                                         stddev=1e-1), name='weights')
            fc1b = tf.Variable(tf.constant(1.0, shape=[5], dtype=tf.float32),
                                 trainable=True, name='biases')
            
            maxpool5_flat = tf.nn.l2_normalize(tf.reshape(self.maxpool5, [-1, int(shape/2)]), 1)
            avgpool5_flat = tf.nn.l2_normalize(tf.reshape(self.avgpool5, [-1, int(shape/2)]), 1)
            
            self.concat = tf.concat([maxpool5_flat, avgpool5_flat], 1)
            self.reg_head = tf.nn.bias_add(tf.matmul(self.concat, fc1w), fc1b, name="reg_val")
            self.parameters += [fc1w, fc1b]
            print("inside dan concat")

    #function to initialize with image-net model
    def initialize_with_imagenet(self, weight_file, sess):
        weights = np.load(weight_file)
        keys = sorted(weights.keys())
        for i, k in enumerate(keys):
            if i==len(self.parameters)-2:
                break
            sess.run(self.parameters[i].assign(weights[k]))
    
    #function to initialize with vgg-face model
    def initialize_with_vggface(self, weight_file, sess):
        data = loadmat(weight_file)
        layers = data['layers'][0]
        i=0
        for layer in layers:
            name = layer[0]['name'][0][0]
            layer_type = layer[0]['type'][0][0]
            if layer_type=='conv' and name[0:2]!='fc':
                kernel, bias = layer[0]['weights'][0][0]
                sess.run(self.parameters[i].assign(kernel))
                sess.run(self.parameters[i+1].assign(bias.reshape(bias.shape[0])))
                print (name, kernel.shape, bias.shape)
                i+=2
    
    #function too load our trained model
    def load_trained_model(self, pickle_file, sess):
        with open(pickle_file, 'rb') as pfile:
            param=pickle.load(pfile)
        for i in range(len(param)):
            sess.run(self.parameters[i].assign(param[i]))
```

# Helper methods

The following cell contains helper methods to print the remaining time left for training in an epoch and another helper method to create the input queue for datasets to be fed into the neural network.

```python
import tensorflow as tf

#to print remaining time in execution
def printTime(remtime):
	hrs = int(remtime)/3600
	mins = int((remtime/60-hrs*60))
	secs = int(remtime-mins*60-hrs*3600)
	timedisp="Time remaining : "
	if hrs>0:
		timedisp+=str(hrs)+"Hrs "
	if mins>0:
		timedisp+=str(mins)+"Mins "
	timedisp += str(secs)+"Secs" 
	print(timedisp)


def decode_train(tr_serialized_example):
    """
      Parses an image and label from the given `serialized_example`.
      It is used as a map function for `dataset.map`
    """  
       
    # Decode the record read by the reader
    tr_feature = {'train/image': tf.FixedLenFeature([], tf.string), 'train/label': tf.FixedLenFeature([], tf.string)}
    tr_features = tf.parse_single_example(tr_serialized_example, features=tr_feature)

    # Convert the image data from string back to the numbers
    tr_image = tf.decode_raw(tr_features['train/image'], tf.uint8)
    tr_label = tf.decode_raw(tr_features['train/label'], tf.float32)

    # Reshape image data into the original shape
    tr_image = tf.reshape(tr_image, [224, 224, 3])
    tr_label = tf.reshape(tr_label, [5])
    
    return tr_image, tr_label

def decode_val(val_serialized_example):
    """
      Parses an image and label from the given `serialized_example`.
      It is used as a map function for `dataset.map`
    """     
    # Decode the record read by the reader
    val_feature = {'val/image': tf.FixedLenFeature([], tf.string), 'val/label': tf.FixedLenFeature([], tf.string)}
    val_features = tf.parse_single_example(val_serialized_example, features=val_feature)
    
    # Convert the image data from string back to the numbers
    val_image = tf.decode_raw(val_features['val/image'], tf.uint8)
    val_label = tf.decode_raw(val_features['val/label'], tf.float32)
    
    # Reshape image data into the original shape
    val_image = tf.reshape(val_image, [224, 224, 3])
    val_label = tf.reshape(val_label, [5])
    
    return val_image, val_label

def decode_test(test_serialized_example):
    """
      Parses an image and label from the given `serialized_example`.
      It is used as a map function for `dataset.map`
    """  
    # Decode the record read by the reader
    test_feature = {'test/image': tf.FixedLenFeature([], tf.string), 'test/label': tf.FixedLenFeature([], tf.string)}
    test_features = tf.parse_single_example(test_serialized_example, features=test_feature)
    
    # Convert the image data from string back to the numbers
    test_image = tf.decode_raw(test_features['test/image'], tf.uint8)
    test_label = tf.decode_raw(test_features['test/label'], tf.float32)
    
    # Reshape image data into the original shape
    test_image = tf.reshape(test_image, [224, 224, 3])
    test_label = tf.reshape(test_label, [5])
    
    return test_image, test_label
```

# Training for the Facial features part of the model

The following cell creates an input queue of datasets for training the network and subsequently using it for validation and testing. 

1. The DAN network with Adam optimizer and a Mean squared Error objective function is used.

2. The weights and biases are stored 'param_' pickle files and the loss values of training, validation and test for each batch are stored in one list each and finally dumped into pickle files.

3. The accuracy metric used in the reference paper is a distance metric between the predicted values and the ground truth and that is calculated for each epoch and is displayed. 

```python
import tensorflow.compat.v1 as tf
import numpy as np
#from dan import DAN
import time
#from remtime import *
import warnings
import pickle
import sys
import matplotlib.pyplot as plt

tf.disable_v2_behavior()
warnings.filterwarnings("ignore")

#defining parameters and number of images
LEARNING_RATE = 0.001
BATCH_SIZE = 32
N_EPOCHS = 3
REG_PENALTY = 0.0005
PER=0.2
NUM_IMAGES = 580764
NUM_VAL_IMAGES = 194269
NUM_TEST_IMAGES = 193988

sys.stdout = open('outfile', 'w')
print('console printed here')


imgs = tf.placeholder('float', [None, 224, 224, 3], name="image_placeholder")
values = tf.placeholder('float', [None, 5], name="value_placeholder")

config = tf.ConfigProto()
config.gpu_options.allow_growth = True
config.gpu_options.per_process_gpu_memory_fraction = 0.8
with tf.Session(config=config) as sess:
    print('here before DAN')
    model = DAN(imgs, REG_PENALTY=REG_PENALTY, preprocess='vggface')
    output = model.output
    cost = tf.reduce_mean(tf.squared_difference(model.output, values))+ model.cost_reg
    optimizer = tf.train.MomentumOptimizer(learning_rate=LEARNING_RATE,momentum = 0.9).minimize(cost)

    #code to create training dataset input pipeline
    tr_dataset = tf.data.TFRecordDataset(['train_full.tfrecords'])
    tr_dataset = tr_dataset.map(decode_train)
    tr_dataset = ((tr_dataset.shuffle(buffer_size=100)).repeat(2*N_EPOCHS)).batch(BATCH_SIZE)
    
    tr_iterator = tr_dataset.make_one_shot_iterator()
    tr_images, tr_labels = tr_iterator.get_next()
    print('here loaded training record')
    
    #code to create validation dataset input pipeline
    val_dataset = tf.data.TFRecordDataset(['val_full.tfrecords'])
    val_dataset = val_dataset.map(decode_val)
    val_dataset = ((val_dataset.shuffle(buffer_size=100)).repeat(N_EPOCHS)).batch(BATCH_SIZE)
    
    val_iterator = val_dataset.make_one_shot_iterator()
    val_images, val_labels = val_iterator.get_next()
    print('here loaded validation record')
    
    #code to create test dataset input pipeline
    test_dataset = tf.data.TFRecordDataset(['test_full.tfrecords'])
    test_dataset = test_dataset.map(decode_test)
    test_dataset = ((test_dataset.shuffle(buffer_size=100)).repeat(2*N_EPOCHS)).batch(BATCH_SIZE)
    
    test_iterator = test_dataset.make_one_shot_iterator()
    test_images, test_labels = test_iterator.get_next()
    print('here loaded test record')
    
    init_op = tf.group(tf.global_variables_initializer(), tf.local_variables_initializer())
    sess.run(init_op)
    print('init session called')
    
    model.initialize_with_vggface('vgg-face.mat', sess)
    loss_list=[]
    val_loss_list=[]
    test_loss_list=[]
    param_num = 1 
    
    for epoch in range(N_EPOCHS):
        tr_acc_list = []
        val_acc_list=[]
        test_acc_list=[]
        test_predict =[]
        sess.run(tf.local_variables_initializer())
        
        #debug variables and for remaning time calculation 
        i=0
        error=0
        stime = time.time()
        z=0
        
        #loading training images and training the model
        while i<NUM_IMAGES:
            if(i == 580736):
                i+=28
            else:
                i+=BATCH_SIZE
                    
            z=z+1
            try:
                epoch_x, epoch_y = sess.run([tr_images, tr_labels])
            except:
                print (error, ": Error in reading this batch")
                error+=1
                if error>10:
                    break
                continue
            _, c = sess.run([optimizer, cost], feed_dict = {imgs: epoch_x.astype(np.float32), values: epoch_y})
            loss_list.append(np.power(c,0.5))
          
            
            x=100/PER
            if not i%2000:
                per = float(i)/NUM_IMAGES*100
                print("Epoch:"+str(round(per,2))+"% Of "+str(epoch+1)+"/"+str(N_EPOCHS)+", Batch loss:"+str(round(c,4)))
                ftime = time.time()
                remtime = (ftime-stime)*((NUM_IMAGES-i)/(NUM_IMAGES/x))
                stime=ftime
                #remtime_list.append(remtime)
                printTime(remtime)
            if not i%20000:
                with open('param'+str(param_num)+'.pkl', 'wb') as pfile:
                    pickle.dump(sess.run(model.parameters), pfile, pickle.HIGHEST_PROTOCOL)
                print (str(param_num)+" weights Saved!!")
                param_num+=1

        with open('param'+str(param_num)+'.pkl', 'wb') as pfile:
            pickle.dump(sess.run(model.parameters), pfile, pickle.HIGHEST_PROTOCOL)
            print (str(param_num)+" weights Saved!!")
            param_num+=1


        sess.run(tf.local_variables_initializer())
        print("Computing Training Accuracy..")
        i=0
        error=0
        while i<NUM_IMAGES:
            if(i == 580736):
                i+=28
            else:
                i+=BATCH_SIZE
                
            try:
                #if(i>500000):
                    #print("Y Labels",i, ":",epoch_y)
                    
                epoch_x, epoch_y = sess.run([tr_images, tr_labels])
            except:
                print ("i:",i,"Error in reading this batch")
                error+=1
                if error>10:
                    break
                continue
            output = sess.run([model.output], feed_dict = {imgs: epoch_x.astype(np.float32)})
            tr_mean_acc = np.mean(1-np.absolute(output-epoch_y))
            tr_acc_list.append(tr_mean_acc)
            
        tr_mean_acc = np.mean(tr_acc_list)

        print("Computing Validation Accuracy..")
        i=0
        error=0
        while i<NUM_VAL_IMAGES:
            if(i == 194240):
                i+=29
            else:
                i+=BATCH_SIZE
                
            try:
                epoch_x, epoch_y = sess.run([val_images, val_labels])
            except:
                print ("i:",i,"Error in reading this batch")
                error+=1
                if error>10:
                    break
                continue
            output = sess.run([model.output], feed_dict = {imgs: epoch_x.astype(np.float32)})          
            val_mean_acc = np.mean(1-np.absolute(output-epoch_y))
            val_acc_list.append(val_mean_acc)
            
            #computing validation loss
            cost1 = np.mean(np.power(np.absolute(np.subtract(output,epoch_y)),2))
            val_loss_list.append(cost1)
            
        val_mean_acc = np.mean(val_acc_list)
        
        print("Computing Test Accuracy..")
        i=0
        error=0
        while i<NUM_TEST_IMAGES:
            if(i == 193984):
                i+=4
            else:
                i+=BATCH_SIZE
                
            try:
                epoch_x, epoch_y = sess.run([test_images, test_labels])            
            except:
                print ("i:",i,"Error in reading this batch")
                error+=1
                if error>10:
                    break
                continue
            output = sess.run([model.output], feed_dict = {imgs: epoch_x.astype(np.float32)})
            test_mean_acc = np.mean(1-np.absolute(output-epoch_y))
            test_acc_list.append(test_mean_acc)
            
            #computing validation loss
            cost2 = np.mean(np.power(np.absolute(np.subtract(output,epoch_y)),2))
            test_loss_list.append(cost2)
            
            #predicted personality scores for test images stored for ensemble
            test_predict.append(output)
            
        test_mean_acc = np.mean(test_acc_list)
        
        print("Epoch"+ str(epoch+1)+" completed out of "+str(N_EPOCHS))
        print("Tr. Mean Acc:"+str(round(tr_mean_acc,4)))
        print("Val. Mean Acc:"+str(round(val_mean_acc,4)))
        print("Test. Mean Acc:"+str(round(test_mean_acc,4)))
    
    #saving different loss lists and prediction scores 
    saver = tf.train.Saver()
    saver.save(sess, 'model_full')
    print ("Session Saved!!")
    with open('loss_full.pkl', 'wb') as pfile:
        pickle.dump(loss_list, pfile, pickle.HIGHEST_PROTOCOL)
    print ("Loss List Saved!!")
    print("saving val_loss_list")
    with open('val_loss_full.pkl', 'wb') as pfile:
        pickle.dump(val_loss_list, pfile, pickle.HIGHEST_PROTOCOL)
    print ("Val Loss List Saved!!")
    print("saving test_loss_list")
    with open('test_loss_full.pkl', 'wb') as pfile:
        pickle.dump(test_loss_list, pfile, pickle.HIGHEST_PROTOCOL)
    print ("test Loss List Saved!!")
    with open('face_predict.pkl', 'wb') as pfile:
        pickle.dump(test_predict, pfile, pickle.HIGHEST_PROTOCOL)
    print ("test prediction Saved!!")
```

# Gaze features

The features for gaze modality are obtained in multiple steps. 

- First the videos are passed through open face tool and gaze features such as raw gaze coordinates, the eye landmarks and gaze angle are extracted for each frame of the video and this is made available by the tool in a 'Videoname.csv' file.

- In the next step each 3 main .csv files, one each for training, test and validation are curated.
    1. Each video's .csv output from OpenFace is loaded into a pandas dataframe with only the necessary colums(x,y coordinates for both eyes, eye landmarks 23, 25, 27, 50, 53, 55)
    2. The top 100 values with maximum confidence(for Openface) are chosen and used for further processing.
    3. The pupil diameter is obatined for each frame and stored in a separate list.
    4. The statistics for the right eye's values of x coordinate, y-coordinate, pupil diameter such as mean, median, minimum, maximum and standard deviation are computed and stored in another .csv file.
    These are gaze features which will be used to train the gaze part of the model for Personality Trait prediction.
    
- A simple regression network is then trained for the gaze features which will predict the personality scores for the gaze modality.

# Pupil Diameter Calculation

The following cell defines a function that is used to calculate pupil diameter from the eye landmarks that are passed as arguments to the function.

```python
# to find pupil diameter
from math import sqrt 
  
def findPupilDiameter(x1, y1, x2, y2, x3, y3) : 
    x12 = x1 - x2;  
    x13 = x1 - x3;  
  
    y12 = y1 - y2;  
    y13 = y1 - y3;  
  
    y31 = y3 - y1;  
    y21 = y2 - y1;  
  
    x31 = x3 - x1;  
    x21 = x2 - x1;  
  
    # x1^2 - x3^2  
    sx13 = pow(x1, 2) - pow(x3, 2);  
  
    # y1^2 - y3^2  
    sy13 = pow(y1, 2) - pow(y3, 2);  
  
    sx21 = pow(x2, 2) - pow(x1, 2);  
    sy21 = pow(y2, 2) - pow(y1, 2);  
  
    f = (((sx13) * (x12) + (sy13) * 
          (x12) + (sx21) * (x13) + 
          (sy21) * (x13)) // (2 * 
          ((y31) * (x12) - (y21) * (x13)))); 
              
    g = (((sx13) * (y12) + (sy13) * (y12) + 
          (sx21) * (y13) + (sy21) * (y13)) // 
          (2 * ((x31) * (y12) - (x21) * (y13))));  
  
    c = (-pow(x1, 2) - pow(y1, 2) - 
         2 * g * x1 - 2 * f * y1);  
  
    # eqn of circle be x^2 + y^2 + 2*g*x + 2*f*y + c = 0  
    # where centre is (h = -g, k = -f) and  
    # radius r as r^2 = h^2 + k^2 - c  
    h = -g;  
    k = -f;  
    sqr_of_r = h * h + k * k - c;  
  
    # r is the radius  
    r = round(sqrt(sqr_of_r), 5);  
  
    #print("Centre = (", h, ", ", k, ")");  
    #print("Radius = ", r);  
    
    return 2*r
```

# Architecture and Training of gaze part of model 

The gaze part employs a simple regression network whose input is structural data fed from a .csv file. 

```python
import numpy as np
import pandas as pd

import tensorflow as tf
from tensorflow import keras

from tensorflow import feature_column
from tensorflow.keras import layers

import matplotlib.pyplot as plt

import sys

np.set_printoptions(threshold=sys.maxsize)

sys.stdout = open('outfile_A', 'w')
print('console printed here')

check = []

# loaad csv data into a pandas frame for further processing
dataframe_val = pd.read_csv('validation_gaze_features_1_N.csv')

dataframe_test = pd.read_csv('test_gaze_features_1_N.csv')

dataframe = pd.read_csv('training_gaze_features_6k_1_N.csv')

print(len(dataframe), 'train examples')
print(len(dataframe_val), 'validation examples')
print(len(dataframe_test), 'test examples')

# A utility method to create a tf.data dataset from a Pandas Dataframe(Input pipeline)
def df_to_dataset(dataframe, shuffle=True, batch_size=32):
    dataframe = dataframe.copy()
    dataframe = dataframe.drop(['Unnamed: 0'],axis=1)
    #labels = dataframe.pop('E')
    #pd.DataFrame([df.pop(x) for x in ['E','A','C','N','O']]).T
    labels = dataframe[['E','A','C','N','O']]
    ds = tf.data.Dataset.from_tensor_slices((dict(dataframe), labels.to_numpy()))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(dataframe))
    ds = ds.batch(batch_size)
    return ds

feature_columns = []

# numeric cols
for header in ['min_dr', 'max_dr', 'mean_dr', 'median_dr', 'std_dr', 
               'min_xr', 'max_xr', 'mean_xr', 'median_xr', 'std_xr', 
               'min_yr', 'max_yr', 'mean_yr', 'median_yr', 'std_yr']:
    feature_columns.append(feature_column.numeric_column(header))
    
feature_layer = tf.keras.layers.DenseFeatures(feature_columns)

batch_size = 32

# invoke method for input pipeline
train_ds = df_to_dataset(dataframe, batch_size=batch_size)
val_ds = df_to_dataset(dataframe_val, shuffle=False, batch_size=batch_size)
test_ds = df_to_dataset(dataframe_test, shuffle=False, batch_size=batch_size)

print(type(train_ds))

# Custom callback function for computation of mean accuracy metric
class Metrics(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self._data = []

    def on_epoch_end(self, batch, logs={}):    
        X_val = val_ds
        X_train = train_ds
        y_predict_val = np.asarray(model.predict(X_val))
        y_predict_train = np.asarray(model.predict(X_train))
        
        print('We are storing predictions with each epoch')

        self._data.append({
            'val_pred':y_predict_val,
            'train_pred':y_predict_train,})
        return

    def get_data(self):
        return self._data

#defining model with input layer, 2 hidden layers and one output layer
model = tf.keras.Sequential([
  feature_layer,
  layers.Dense(30, activation='linear'),
  layers.Dropout(0.1),
  layers.Dense(30, activation='relu'),
  layers.Dense(5, activation='sigmoid')
])

opt = keras.optimizers.RMSprop(learning_rate=0.0005)

metric = Metrics()
model.compile(optimizer=opt,
              loss='mean_squared_error',
              metrics=['mean_absolute_error'])

#invoking model with number of epochs to train and callbacks
history = model.fit(train_ds,
          validation_data=val_ds,
          epochs=75,callbacks =[metric])

check = metric.get_data()

# Plot training & validation loss values
plt.figure(num=3, figsize=(6, 4), dpi=120, facecolor='w', edgecolor='r')
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
#plt.yticks(np.arange(0.0200, 0.0250, step=0.0010))
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()
```

# Ensembling - Late Fusion

The scores of each frame from facial features modality are averaged to give the values for each video. The values from facial features modality and gaze modality are averaged with equal weightage to give the final prediction score.  Accuracy is also computed for each trait.

```python
import os, os.path
import sys
import pandas as pd
import pickle

sys.stdout = open('outfile_E', 'w')
print('console printed here')

df = pd.read_csv('./data/test_gt.csv')
NUM_VID = len(df)
labels = []
eye_gaze =[]

list1 = []
listImg = []

#obtaining list of number of images extracted per video
for j in range(NUM_VID):
    DIR='./ImageData/testData/'+(df['VideoName'].iloc[j]).split('.mp4')[0] 
    list1 = ([name for name in os.listdir(DIR) if os.path.isfile(os.path.join(DIR, name))])
    listImg.append(len(list1))
    
print(listImg)
print(len(listImg))

with open('listImg.pkl', 'wb') as pfile:
    pickle.dump(listImg, pfile, pickle.HIGHEST_PROTOCOL)
print ("listImg Saved!!")
```

```python
import pickle
import sys
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
np.set_printoptions(threshold=sys.maxsize)

sys.stdout = open('outfile_en', 'w')
print('console printed here')

#loading gaze predictions
with open('gaze_predict.pkl', 'rb') as f:
    gaze_predict = pickle.load(f, encoding='latin1')

print(type(gaze_predict))
print(gaze_predict.shape)
#print(gaze_predict)

#loading face predictions
with open('face_predict.pkl', 'rb') as f:
    pre_face = pickle.load(f, encoding='latin1')
    
print(type(pre_face))
print(len(pre_face))
#print(pre_face)

#loading list of number of images extracted per video 
with open('listImg.pkl', 'rb') as f:
    listNumImg = pickle.load(f, encoding='latin1')
    
print(type(listNumImg))
print(len(listNumImg))
#print(listNumImg)
print(sum(listNumImg))

#computing the predictions for one video from facial features modality by averaging values from induvidual frames
ff_predict = np.empty((193988,5))
                         
for i in range(len(pre_face)):
    juju = np.reshape(np.asarray(pre_face[i]), (32,5))
        
    ff_predict= np.append(ff_predict,juju,axis = 0)

ff_predict = ff_predict[193988:387976,:]
print(ff_predict.shape)
#print(ff_predict)

k = 0
avg = []
face_predict = []

for i in range(len(listNumImg)):
    iteration = ff_predict[k:k+listNumImg[i],:]
    #print(iteration)
    avg = np.sum(iteration, axis=0)/listNumImg[i]
        
    face_predict.append(avg)
    k = k+listNumImg[i]
    
face_predict = np.asarray(face_predict)
print(face_predict.shape)
#print(face_predict)

# ensemble of prediction scores of both modalities
ensemble = np.add(1*gaze_predict,1*face_predict)
#print(ensemble)
ensemble = ensemble/2
#print(ensemble)

df = pd.read_csv('test_gaze_features_2k_15.csv')
NUM_VID = len(df)
labels = []

y_label_test = []

for i in range(NUM_VID):    
    labels = df[['E','A','C','N','O']].iloc[i]
    y_label_test = np.append(y_label_test, labels.to_numpy(), axis = 0)
    
y_label_test = np.reshape(y_label_test,(2000,5))
    
ensemble_mean_acc = np.mean(1-np.absolute(ensemble-y_label_test))
face_test_acc  = np.mean(1-np.absolute(face_predict-y_label_test))

print('ensemble overall')
print(ensemble_mean_acc)

print('Face mean accuracy')
print(face_test_acc)

print('just face')
e_mean_acc_f = np.mean(1-np.absolute(face_predict[:,0]-y_label_test[:,0]))
print('E')
print(e_mean_acc_f)
a_mean_acc_f = np.mean(1-np.absolute(face_predict[:,1]-y_label_test[:,1]))
print('A')
print(a_mean_acc_f)
n_mean_acc_f = np.mean(1-np.absolute(face_predict[:,2]-y_label_test[:,2]))
print('N')
print(n_mean_acc_f)
c_mean_acc_f = np.mean(1-np.absolute(face_predict[:,3]-y_label_test[:,3]))
print('C')
print(c_mean_acc_f)
o_mean_acc_f = np.mean(1-np.absolute(face_predict[:,4]-y_label_test[:,4]))
print('O')
print(o_mean_acc_f)

print('ensemble')
e_mean_acc_e = np.mean(1-np.absolute(ensemble[:,0]-y_label_test[:,0]))
print('E')
print(e_mean_acc_e)
a_mean_acc_e = np.mean(1-np.absolute(ensemble[:,1]-y_label_test[:,1]))
print('A')
print(a_mean_acc_e)
n_mean_acc_e = np.mean(1-np.absolute(ensemble[:,2]-y_label_test[:,2]))
print('N')
print(n_mean_acc_e)
c_mean_acc_e = np.mean(1-np.absolute(ensemble[:,3]-y_label_test[:,3]))
print('C')
print(c_mean_acc_e)
o_mean_acc_e = np.mean(1-np.absolute(ensemble[:,4]-y_label_test[:,4]))
print('O')
print(o_mean_acc_e)
```

# P- value for statistical significance

The statistical significance of adding gaze modality is checked here. 
Null hypothesis - Gaze doesn't influence Personality prediction
Alternative hypothesis - Gaze influences Personality prediction

```python
sys.stdout = open('outfile_p', 'w')
print('console printed here')

# computing chisquare value for ensemble (Alternative hypothesis)
chisquare = np.square(ensemble - y_label_test)
for i in range(2000):
    for j in range(5):
        chisquare[i,j] = chisquare[i,j]/y_label_test[i,j]
#print(chisquare)

# computing chisquare value for face only (Null hypothesis)
chisquare_f = np.square(face_predict - y_label_test)
for i in range(2000):
    for j in range(5):
        chisquare_f[i,j] = chisquare_f[i,j]/y_label_test[i,j]
#print(chisquare_f)

# computing overall p-score to read off exact p-value from table(Null hypothesis)
p_score = chisquare_f+chisquare
#print(p_score)
print(np.sum(p_score, axis=0)/2000)
```
